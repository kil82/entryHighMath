```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,4*np.pi,0.1)
y1 = np.cos(x)
y2 = np.cos(2*x)
```

```python
plt.plot(x,y1)
plt.plot(x, y2)
plt.legend(['cos(x)', 'cos(2*x)'])
plt.show()
```
